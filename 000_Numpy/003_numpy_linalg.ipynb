#### linalg.norm(x, ord=None, axis=None, keepdims=False)
- 이 함수는 ord 매개변수 값에 따라 8개의 서로 다른 행렬의 norm중 하나를 반환하거나,
- 무한수의 벡터 Norm을 반환할 수 있다.
  
x : 입력 배열. axis=None 이면, x는 ord가 None이 아닌 한 x는 1, 2차원이어야 한다.  
axis=None, ord=None 이면 x.ravel의 2-norm이 반환된다.  
  
ord : (아래 표를 참고한다)  
  
axis : axis가 정수인 경우, 벡터 norm을 계산할 x 축을 지정한다. 축이 (2-tuple)인 경우, 2차원 행렬 축을 지정하고 행렬 norm이 계산된다. 축이 None이면 벡터 Norm(x가 1-D인 경우), 행렬 Norm(x가 2-D인 경우)가 반환된다.

```python
import numpy as np
a = (np.arange(9) - 4)
b = a.reshape((3, 3))
print(a)
print(b)

print()

print(np.linalg.norm(a))
print(np.linalg.norm(b))

print()

print(np.linalg.norm(a, np.inf)) # <-- max(abs(a))
print(np.linalg.norm(b, np.inf)) # <-- max(sum(abs(x), axis=1))
'''
[[9],
 [2],
 [9]] 중에서 최대 값
'''
print()

print(np.linalg.norm(a, -np.inf)) # <-- min(abs(a))
print(np.linalg.norm(b, -np.inf)) # <-- min(sum(abs(x), axis=1))
'''
[[9],
 [2],
 [9]] 중에서 최소 값
'''
print()

print(np.linalg.norm(a, 1)) # <-- sum(abs(x)**ord)**(1./ord)
'''
[20]
'''
print(np.linalg.norm(b, 1)) # <-- max(sum(abs(x), axis=0))
'''
[7, 6, 7] 중에서 최대 값
'''


```

```python
a = (np.arange(9) - 4)
b = a.reshape((3, 3))

print(np.linalg.norm(a, 2)) # <-- sum(abs(x)**ord)**(1./ord)
'''
[58]^(1/2)
'''
print(np.linalg.norm(b, 2)) # <-- 2-norm (largest sing. value)
'''
[7, 6, 7] 중에서 최대 값
'''

print()

a = np.array([1, 2, 3, 4])
print(np.linalg.norm(a, 2))

b = np.array([[1, 2],
              [3, 4]])
print(np.linalg.norm(b, 2)) # 왜 값이 달라지는지??
```

```python
b = a.reshape((3, 3))

print(np.linalg.norm(b, axis=0))
print(np.linalg.norm(b, axis=1))
```

### (참고) 두 점사이 거리 구하는 3가지 방법

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

ax.scatter(0, 0, 0)
ax.scatter(3, 3, 3)
plt.show()
```

```python
point_1 = np.array((0, 0, 0))
point_2 = np.array((3, 3, 3))
square = np.square(point_1 - point_2)
sum_square = np.sum(square)
distance = np.sqrt(sum_square)
# distance = np.sqrt(np.sum(np.square(point_1 - point_2)))
# distance = np.sqrt(sum(pow(a-b, 2) for a, b in zip(point_1, point_2)))
print(distance)
```

```python
# 디폴트 값으로 Euclidean distance 거리 반환, L2-norm, Frobenius norm
print(np.linalg.norm(point_1 - point_2))
```

```python
# Take the difference between the 2 points
diff = point_1 - point_2
# Perform the dot product on the point with itself to get the sum of the squares
sum_square = np.dot(diff, diff) # 같은 벡터 끼리 내적하면 길이 제곱 된다.
# Get the square root of the result
distance = np.sqrt(sum_square)
print(distance)
```
