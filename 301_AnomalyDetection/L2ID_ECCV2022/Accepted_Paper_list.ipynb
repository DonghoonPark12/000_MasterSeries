#### One-shot learning vs Few-shot learning vs Zero-shot learning

#### Inception Score(IS)
- IS는 GAN의 성능평가에 두 가지 기준을 사용
- ✓ 생성된 영상의 품질
- ✓ 생성된 영상의 다양성(diversity)
- IS는 Inception모델에서 식별하기 쉬운 영상 및 식별된 레이블의 Variation(편차, 변화)이 풍부할수록 score가 높게 출력 되도록 설계


#### LPIPS
- 비교할 2개의 이미지를 각각 VGG Network에 넣고, 중간 layer의 feature값들을 각각 뽑아내서, 2개의 feature가 유사한지를 측정하여 평가지표로 사용
- 두 이미지 $x, x_0$ 가 주어졌을 때, ImageNet으로 pretrain된 network의 몇몇 layer 에서의 activation map $\hat{y^l_{hw}}, \hat{y^l_{0hw}}$을 얻어 Euclidean distance 계산한 후 $w^l$로 scaling한 다음 channel-wise averaging을 한 값을 에 대해 평균을 내면 LPIPS가 얻어집니다
   
- 기존의 IS나 FID와는 다르게 유사도를 사람의 인식에 기반하여 측정하려 시도

$$ LPIPS=\sum_l{\cfrac{1}{H_lW_l}}\sum_{h,w}{\Vert{w^l{\odot}(\hat{y}^l_{hw}-\hat{y}^l_{0hw})}\Vert}^2_2 $$
