### App model summary

- Streamlit 앱은 위에서 아래로 실행되는 Python 스크립트입니다.
- 사용자가 `앱을 가리키는 브라우저 탭을 열 때마다 스크립트가 실행되고 새 세션이 시작`됩니다.
- 스크립트가 실행되면 Streamlit은 브라우저에 실시간으로 출력을 그립니다.
- `사용자가 위젯과 상호 작용할 때마다 스크립트가 다시 실행되고 Streamlit은 브라우저에서 해당 출력을 다시 그립니다`.
- `해당 위젯의 출력 값은 재실행 중 새 값과 일치합니다`.
- 스크립트는 비용이 많이 드는 함수를 다시 계산하지 않기 위해 `Streamlit 캐시`를 사용하므로 업데이트가 매우 빠르게 이루어집니다.
- `세션 상태를 사용하면 간단한 위젯 이상이 필요할 때 재실행 간에 유지되는 정보를 저장할 수 있습니다`.
- Streamlit 앱은 페이지 폴더에 별도의 .py 파일로 정의된 여러 페이지를 포함할 수 있습니다.
